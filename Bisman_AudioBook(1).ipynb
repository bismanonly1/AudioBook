```python
import PyPDF2
import re
```

```python
from google.colab import files
upload = files.upload()
```

```python
pdf = list(upload.keys())[0]
def extract_text_from_pdf(pdf):
    reader = PyPDF2.PdfReader(pdf)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text
text = extract_text_from_pdf(pdf)
def clean_text(text):
    import re
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
print("Extracted Text from PDF:")

cleaned_text = clean_text(text)

print("\nCleaned Text:")
print(cleaned_text)
```

```python
import pyttsx3
```

```python
def initialize_speaker():
    speaker = pyttsx3.init()

    rate = speaker.getProperty('rate')
    speaker.setProperty('rate', rate - 50)

    volume = speaker.getProperty('volume')
    speaker.setProperty('volume', 1)

    voices = speaker.getProperty('voices')
    speaker.setProperty('voice', voices[1].id)

    return speaker
```

```python
def text_to_speech(speaker, text, output_audio_path):
    speaker.save_to_file(text, output_audio_path)
    speaker.runAndWait()
```

```python
speaker = initialize_speaker()
text = "Hello, this is an example of text-to-speech conversion."
output_audio_path = r"D:\gbc\ML2\output_audio.mp3"
text_to_speech(speaker, text, output_audio_path)

print(f"Audio saved to {output_audio_path}")
```
